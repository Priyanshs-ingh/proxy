```python
import requests
import json
from datetime import datetime

# ============= INPUT YOUR CREDENTIALS HERE =============
PROXYCURL_API_KEY = "YOUR_API_KEY"  # Replace with your actual API key
LINKEDIN_PROFILE_URL = "https://www.linkedin.com/in/dr-gautam-agrawal-aa45b541"  # Replace with the LinkedIn URL you want to scrape
# ====================================================

def format_date(date_dict):
    if not date_dict:
        return "Present"
    month = str(date_dict.get('month', '')).zfill(2)
    year = str(date_dict.get('year', ''))
    return f"{month}/{year}" if month and year else "N/A"

def get_linkedin_data():
    api_endpoint = 'https://nubela.co/proxycurl/api/v2/linkedin'
    headers = {'Authorization': f'Bearer {PROXYCURL_API_KEY}'}

    params = {
        'url': LINKEDIN_PROFILE_URL,
        'fallback_to_cache': 'on-error',
        'use_cache': 'if-present',
        'skills': 'include',
        'inferred_salary': 'include',
        'personal_email': 'include',
        'personal_contact_number': 'include',
        'twitter_profile_id': 'include',
        'facebook_profile_id': 'include',
        'github_profile_id': 'include'
    }

    try:
        response = requests.get(
            url=api_endpoint,
            params=params,
            headers=headers,
            timeout=30
        )
        response.raise_for_status()
        data = response.json()

        # Base profile data
        profile_data = {
            'full_name': data.get('full_name', ''),
            'headline': data.get('headline', ''),
            'summary': data.get('summary', ''),
            'country': data.get('country_full_name', ''),
            'email': data.get('personal_email', ''),
            'github': data.get('github_profile_id', ''),
            'twitter': data.get('twitter_profile_id', ''),
            'experiences': [],
            'clubs': []
        }

        # Process experiences for work and clubs
        experiences = data.get('experiences', [])
        leadership_keywords = ["President", "Vice President", "Secretary", "Treasurer"]
        for exp in experiences:
            experience = {
                'title': exp.get('title', ''),
                'company': exp.get('company', ''),
                'company_linkedin_url': exp.get('company_linkedin_profile_url', ''),
                'duration': f"{format_date(exp.get('starts_at'))} - {format_date(exp.get('ends_at'))}",
                'location': exp.get('location', ''),
                'description': exp.get('description', '')
            }
            profile_data['experiences'].append(experience)

            # Filter for club-related roles
            if any(keyword.lower() in exp.get('title', '').lower() for keyword in leadership_keywords):
                profile_data['clubs'].append(experience)

        # Save data to JSON
        timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')
        filename = f'linkedin_profile_{timestamp}.json'

        with open(filename, 'w', encoding='utf-8') as json_file:
            json.dump(profile_data, json_file, indent=4, ensure_ascii=False)
        
        print(f"Data successfully saved to {filename}")
        return profile_data

    except requests.exceptions.Timeout:
        print("Request timed out. Please check your network connection or try again later.")
    except requests.exceptions.HTTPError as errh:
        print(f"HTTP Error: {errh}")
        if response.status_code == 403:
            print("Authentication failed. Please check your API key.")
        elif response.status_code == 429:
            print("Rate limit exceeded. Please wait before making another request.")
        else:
            print(f"Proxycurl Error: {response.json().get('error', 'No error message available')}")
    except requests.exceptions.ConnectionError as errc:
        print(f"Error Connecting: {errc}")
    except requests.exceptions.RequestException as err:
        print(f"An error occurred: {err}")
    except json.JSONDecodeError:
        print("Error decoding the API response")
    return None

if __name__ == "__main__":
    profile_data = get_linkedin_data()
    if profile_data:
        print("Profile data fetched successfully!")
```
